# Feature 1 – The chat parser

```python
filename = "hostel_bois.txt"

with open(filename, "r", encoding="utf-8") as file:
    lines = file.readlines()

print("Chat file loaded successfully.")
print("Total lines:", len(lines))
```

# Initialize Variables

```python
messages = []

system_messages = 0
media_messages = 0
deleted_messages = 0
```

# Parse the Chat

Each valid message is separated into:

- Timestamp
- Sender
- Message

Invalid/system messages are skipped.

```python
for line in lines:

    line = line.strip()

    if line == "":
        continue

    if " - " not in line:
        continue

    first = line.split(" - ", 1)

    if len(first) != 2:
        continue

    timestamp = first[0]
    remaining = first[1]

    if ": " not in remaining:
        system_messages += 1
        continue

    sender, text = remaining.split(": ", 1)

    if text == "<Media omitted>":
        media_messages += 1

    if text == "This message was deleted":
        deleted_messages += 1

    messages.append({
        "timestamp": timestamp,
        "sender": sender.strip(),
        "text": text.strip()
    })
```

Display statistics after parsing.

# Feature 4 – NumPy Activity Heatmap



```python
people = [
    "Rahul",
    "Priya",
    "Aman",
    "Karan",
    "Neha",
    "Vikas"
]

person_index = {}

for i in range(len(people)):
    person_index[people[i]] = i

heatmap = np.zeros((len(people), 24), dtype=int)
```

# Fill the Heatmap Matrix

```python
for msg in messages:

    sender = msg["sender"]

    if sender not in person_index:
        continue

    hour = int(msg["timestamp"].split(", ")[1].split(":")[0])

    row = person_index[sender]

    heatmap[row][hour] += 1
```

# Display Heatmap Matrix

```python
print("Activity Heatmap Matrix\n")

for i in range(len(people)):
    print(f"{people[i]:<10}", heatmap[i])
```

# Text-Based Activity Heatmap

Visualize activity using symbols:
- `.` = No activity
- `░` = Low activity
- `▒` = Medium activity
- `█` = High activity

```python
print("TEXT HEATMAP\n")

print("      ", end="")

for h in range(24):
    print(f"{h:02d}", end=" ")

print()

for i in range(len(people)):

    print(f"{people[i]:<7}", end=" ")

    maximum = max(heatmap[i])

    for value in heatmap[i]:

        if maximum == 0:
            symbol = "."
        else:
            ratio = value / maximum

            if ratio == 0:
                symbol = "."
            elif ratio <= 0.25:
                symbol = "░"
            elif ratio <= 0.50:
                symbol = "▒"
            else:
                symbol = "█"

        print(symbol, end="  ")

    print()
```

# Feature 5 – Top Words Analysis

This feature identifies the most frequently used words in the WhatsApp chat after removing common stop words.

```python
stop_words = {
    "the", "is", "am", "are", "was", "were",
    "a", "an", "to", "of", "in", "on",
    "at", "for", "and", "or", "if", "but",
    "i", "you", "he", "she", "it", "we",
    "they", "me", "my", "your", "our",
    "this", "that", "these", "those"
}
```

# Create Word Frequency Dictionary

Store the frequency of each meaningful word.

```python
word_count = {}

for msg in messages:

    text = msg["text"].lower()

    words = text.split()

    for word in words:

        word = word.strip(".,!?()[]{}'\":;")

        if word == "":
            continue

        if word in stop_words:
            continue

        if word not in word_count:
            word_count[word] = 0

        word_count[word] += 1
```
